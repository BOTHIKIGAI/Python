```python
import matplotlib.pyplot as plt
```

```python
apps = ["Tiktok", "WhatsApp", "Facebook", "Instagram", "Zoom"]
downloads = [850, 600, 540, 503, 477]

plt.bar(apps, downloads)
plt.xticks(rotation=70, fontsize=12)
plt.xlabel("App")
plt.ylabel("Numero de descargas (Millones)")
plt.title("Apps más descargadas en el 2020")
plt.show()
```

```python
candidato = ["Sadiq Khan","shaum Bailey","Sian Berry","Luisa Porritt","Niko Omilana","Laurence Fox"][::-1]
votos = [145263245,15451245,215454321,54512167,12548551,5857841121][::-1]

plt.figure(figsize=(10,5))
plt.barh(candidato,votos)
plt.xticks(rotation=70,fontsize=12)
plt.xlabel("Numero de votos (Millones)")
plt.ylabel("Candidato")
plt.title("Ejercicios alcaldia de londres 2020")
plt.show()
```

```python
app = ["Tiktok","WhatsApp","Facebook",
"Instagram","Zoom"]
downloads = [850,600,540,503,477]
usa_downloads=[89,45,53,62,81]

plt.bar(apps,downloads)
plt.bar(apps,usa_downloads,bottom=downloads)
plt.legend(["Resto del mundo","USA"])
plt.xticks(rotation=70,fontsize=12)
plt.xlabel("App")
plt.ylabel("Numero de descargas (Millones)")
plt.title("Apps más descargadas en el 2020")
plt.show()
```
