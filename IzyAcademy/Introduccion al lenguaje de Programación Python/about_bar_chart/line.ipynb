```python
""" 
Útil para ver cambios de una variable en el tiempo, ejemplo: Stock market price,
visitor to your website, visualizar la tendencia en un valor numérico durante un 
intervalo de tiempo continuo.
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0.001
sigma = 0.01
start_price = 5

np.random.seed(0)
returns = np.random.normal(loc=mu, scale=sigma, size=100)
price = start_price * (1 + returns).cumprod()

plt.plot(price)
plt.title("Precio del petroleo")
plt.xlabel("Día")
plt.ylabel("Precio")
```
