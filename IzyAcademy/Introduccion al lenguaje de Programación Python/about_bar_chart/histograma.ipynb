```python
""" 
Un histograma muestra la distribución de datos numericos a traves de un intervalo continuo
mediante la segmentación de datos en diferentes contenedores (Bins) o rangos. Util para
inspeccionar la asimetria en los datos. Util para variables continua ejemplo: Time,
age, weight, height, etc. Se usa mucho por rango. Por defecto, usa 10 bins o contenedores.

Nota: Los histogramas se usan con datos continuos, mientras que los diagramas de barras se
usan con datos categóricos.
"""
```

```python
import matplotlib.pyplot as plt
```

```python
my_data = [163, 163, 164, 170, 170, 172, 173, 190, 190, 180];
sili_data = [161, 172, 174, 175, 181, 183, 186, 190, 180, 185];
bins = [160, 170, 180, 190];
names = ['My friends', 'Sili\'s friend'];

plt.hist([my_data,sili_data],bins=bins, label=names)
plt.title("Mine and Sili's friends height")
plt.xlabel("Number of people")
plt.ylabel("Height in cm")


plt.legend()
plt.show() 
```
